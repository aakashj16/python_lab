## Problem 1: Sum of Two Numbers

Create a function that takes two numbers as arguments and returns their sum.

```python
# sum of two numbers
def sum_two_numbers(a: int | float, b: int | float) -> float | str:
    '''
    The function returns sum of two numbers.

    Arguments:
        a(float)
        b(float)

    Returns:
        Sum of numbers.
    '''

    if isinstance(a, (int, float)) and isinstance(b, (int, float)):
        return a + b

    return 'Please enter valid data.'

num_1, num_2 = 3.14, 5
result = sum_two_numbers(num_1, num_2)

print(f'Sum of {num_1} and {num_2} is {result}')
```

## Problem 2: Convert Mintues to Seconds

Write a function that takes an integer minutes and converts it to seconds.

```python
# minutes to seconds
def minutes_to_seconds(minutes: int) -> int | str:
    '''
    This function converts minutes to seconds.

    Args:
        minutes(int): Time in minutes.

    Returns:
        Time in minutes converted to time in seconds.
    '''

    if isinstance(minutes, int) and minutes >= 0:
        return minutes * 60

    return 'Please enter valid data'

time_minutes = 5
time_seconds = minutes_to_seconds(time_minutes)

print(f'{time_minutes} converted to seconds: {time_seconds}')
```

## Problem 3: Next Number From The Integer Passed

Create a function that takes a number as an argument, increments the number by +1 and returns the result.

```python
# next number from the integer passed
def next_number(number: int) -> int | str:
    '''
    This function returns the next number from the integer passed.

    Args:
        number(int): Integer value.

    Returns:
        Next number from the integer passed.
    '''

    if isinstance(number, int):
        return number + 1

    return 'Please enter valid data.'

num = 0
next_num = next_number(num)

print(f'Integer value next to {num} is: {next_num}')
```

## Proglem 4: Area Of A Triangle

Write a function that takes the base and height of a triangle and returns its area. The area of a triangle is (base * height) / 2

```python
# area of a triangle
def area_of_triangle(base: int | float, height: int | float) -> int | float:
    '''
    This function calculates the area of a triangle.

    Args:
        base (int | float): Base of the triangle
        height (int | float): Height of the triangle

    Returns:
        Area of the triangle
    '''

    if isinstance(base, (int, float)) and isinstance(height, (int, float)) and base > 0 and height > 0:
        return base * height * 0.5
    
    return 'Please enter valid data'

base_triangle = 7
height_triangle = 3
area_triangle = area_of_triangle(base_triangle, height_triangle)
print(f'Area of triangle: {area_triangle}')
```

## Problem 5: Convert Age To Days
Create a function that takes the age in years and returns the age in days. Use 365 days as the length of a year for this challenge.
Ignore leap years and days between last birthday and now.

```python
# convert age to day
def age_to_days(age: int) -> int:
    '''
    This function converts age in years to days.

    Args:
        age(int): Age in years.

    Returns:
        Age in days.
    '''

    if isinstance(age, int) and age > 0:
        return age * 365
    
    return 'Please enter valid data'

age = 33
print(f'I\'m {age} years old. My age in days is {age_to_days(age)} days.')
```

## Problem 6: Cube Of A Number

Write a code to return the cube of a number.

```python
# cube of a number
def cube(a: int | float) -> int | float:
    '''
    This function returns the cube of a number.

    Args:
        a(int | float)

    Returns:
        Cube of a number
    '''

    if isinstance(a, (int, float)):
        return a ** 3

    return 'Please enter valid data'

num = 7
print(f'Cube of {num} is {cube(num)}.')
```

## Problem 7: Return The First Element In An Array

Create a function that takes an array and returns the first element.

```python
# return the first element in an array
def first_element(array: list) -> list[0]:
    '''
    This function returns the first element of an array

    Args:
        array(list)

    Returns:
        First element of the array
    '''
    if array:
        return array[0]

    return 'Please enter valid data'

my_list = ['Sun', 'Mon', 'Tue', 'Wed', 'Thurs', 'Fri', 'Sat']
print(f'First element of the list: {first_element(my_list)}')
```

## Problem 8: Power Calculator

Create a function that takes the voltage and the current, and returns the calculated power. Power is calcuated as voltage times current.

```python
# power calculator
def power_calculator(voltage: int | float, current: int | float) -> int | float:
    '''
    This function calculates power.

    Args:
        voltage(int | float)
        current(int | float)

    Return:
        Power
    '''

    if voltage > 0 and isinstance(voltage, (int, float)) and current > 0 and isinstance(current, (int, float)):
        return voltage * current

    return 'Please enter valid data'

current = 100
voltage = 10
power = power_calculator(voltage, current)
print(f'Current: {current}\nVoltage: {voltage}\nPower: {power}')
```

## Problem 9: Convert Hours Into Seconds

Write a function that converts hours into seconds.

```python
# convert hours into seconds
def hours_to_seconds(hours: int) -> int:
    '''
    This function converts hours into seconds

    Args:
        hours(int)

    Returns:
        Hours converted to seconds
    '''
    if isinstance(hours, int) and hours > 0:
        return hours * 60 * 60
    
    return 'Please enter valid data'

hours = 5
print(f'{hours} hours in seconds equals to: {hours_to_seconds(hours)}')
```

## Problem 10: Maximum Edge Of A Triangle

Create a function that finds the maximum range of a triangle's third edge, where the side lengths are all integers. Maximum range of third edge = Side 1 + Side 2 - 1

```python
# maximum edge of a triangle
def maximum_edge(edge_1: int, edge_2: int) -> int:
    '''
    This function calculates the maximum range of the third edge of a triangle.

    Args:
        edge_1(int)
        edge_2(int)

    Returns:
        Maximum range of the third edge of a triangle
    '''

    if isinstance(edge_1, int) and edge_1 > 0 and isinstance(edge_2, int) and edge_2 > 0:
        return edge_1 + edge_2 - 1
    
    return 'Please enter valid data'

edge_1, edge_2 = 8, 10
print(f'Edge 1 = {edge_1}\nEdge 2 = {edge_2}\n')
print(f'Maximum range of third edge: {maximum_edge(edge_1, edge_2)}')
```

## Problem 11: Remainder Of Two Numbers

Create a function that returns the remainder when first number is divided by the second.

```python
# remainder of two numbers
def calculate_remainder(numerator: int, denominator: int) -> int:
    '''
    This function calculates the remainder of two numbers

    Args:
        Numerator(int)
        Denominator(int)

    Returns:
        Remainder of two numbers
    '''

    if isinstance(numerator, int) and isinstance(denominator, int) and denominator > 0:
        return numerator % denominator
    
    return 'Please enter valid data'

numerator, denominator = 17, 4
print(f'Numerator: {numerator}\nDenominator: {denominator}')
print(f'Remainder: {calculate_remainder(numerator, denominator)}')
```

## Problem 12: Perimeter Of A Rectangle

Create a function that takes length and width and finds the perimeter of a rectangle.

```python
# perimeter of a rectangle
def rectangle_perimeter(length: int | float, breadth : int | float) -> int | float:
    '''
    This function calculates the perimeter of a rectangle.

    Args:
        length(int | float): Length of rectangle in cm
        breadth(int | float): Breadth of rectangle in cm

    Returns:
        Perimeter of rectangle
    '''

    if isinstance(length, (int, float)) and length > 0 and isinstance(breadth, (int, float)) and breadth > 0:
        return 2 * (length + breadth)
    
    return 'Please enter valid data'

length_cm, breadth_cm = 20, 10
print(f'Length of rectange: {length_cm} cm')
print(f'Breadth of rectange: {breadth_cm} cm')
print(f'Perimeter: {rectangle_perimeter(length_cm, breadth_cm)} cm')
```
