## Problem 1: Sum of Two Numbers

Create a function that takes two numbers as arguments and returns their sum.

```python
# sum of two numbers
def sum_two_numbers(a: int | float, b: int | float) -> float | str:
    '''
    The function returns sum of two numbers.

    Arguments:
        a (float)
        b (float)

    Returns:
        float | str:
            Sum of numbers.
    '''

    if isinstance(a, (int, float)) and isinstance(b, (int, float)):
        return a + b

    return 'Please enter valid data.'

num_1, num_2 = 3.14, 5
result = sum_two_numbers(num_1, num_2)

print(f'Sum of {num_1} and {num_2} is {result}')
```

## Problem 10: Maximum Edge Of A Triangle

Create a function that finds the maximum range of a triangle's third edge, where the side lengths are all integers. Maximum range of third edge = Side 1 + Side 2 - 1

```python
# maximum edge of a triangle
def maximum_edge(edge_1: int, edge_2: int) -> int | str:
    '''
    This function calculates the maximum range of the third edge of a triangle.

    Args:
        edge_1 (int)
        edge_2 (int)

    Returns:
        int | str:
            Maximum range of the third edge of a triangle
    '''

    if isinstance(edge_1, int) and edge_1 > 0 and isinstance(edge_2, int) and edge_2 > 0:
        return edge_1 + edge_2 - 1
    
    return 'Please enter valid data'

edge_1, edge_2 = 8, 10
print(f'Edge 1 = {edge_1}\nEdge 2 = {edge_2}\n')
print(f'Maximum range of third edge: {maximum_edge(edge_1, edge_2)}')
```

## Problem 12: Perimeter Of A Rectangle

Create a function that takes length and width and finds the perimeter of a rectangle.

```python
# perimeter of a rectangle
def rectangle_perimeter(length: int | float, breadth : int | float) -> int | float | str:
    '''
    This function calculates the perimeter of a rectangle.

    Args:
        length (int | float): Length of rectangle in cm
        breadth (int | float): Breadth of rectangle in cm

    Returns:
        int | float | str:
            Perimeter of rectangle
    '''

    if isinstance(length, (int, float)) and length > 0 and isinstance(breadth, (int, float)) and breadth > 0:
        return 2 * (length + breadth)
    
    return 'Please enter valid data'

length_cm, breadth_cm = 20, 10
print(f'Length of rectange: {length_cm} cm')
print(f'Breadth of rectange: {breadth_cm} cm')
print(f'Perimeter: {rectangle_perimeter(length_cm, breadth_cm)} cm')
```

## Problem 13: Sum Of Polygon Angles

Given an n-sided regular polygon, return the total sum of internal angles (in degrees). The formula (n - 2) x 180 gives the sum of all the measures of the angles of an n-sided polygon.

```python
# sum of polygon angles
def sum_polygon(num_sides: int) -> int | str:
    '''
    This function calculates the sum of a polygon angles.

    Args:
        num_sides (int): Number of sides of a polygon. 
        Note: num_sides should be greater than 2

    Returns:
        int | str:
            Sum of internal angles (in degrees)
    '''
    if num_sides > 2 and isinstance(num_sides, int):
        return (num_sides - 2) * 180
    
    return 'Please enter valid data.'

number_of_sides = 3
print(f'For a {number_of_sides} sided polygon, the sum of angles is {sum_polygon(number_of_sides)}')
```

Create a function that takes a base number and an exponent number and returns the calculation.

## Problem 15: Football Points

Create a function that takes the number of wins, draws and losses and calculates the number of points a football team has obtained so far.
- Wins get 3 points
- Draws get 1 point
- Losses get 0 points

```python
# football points
def football_points(wins: int = 0, draws: int = 0, losses: int = 0) -> int | str:
    '''
    This function calculates the number of points a football team has obtained so far based on their performance.
    Wins get 3 points each.
    Draws get 1 point each.
    Losses get 0 point each.

    Args:
        wins (int): Number of games a football team has won.
        draws (int): Number of games ended in a draw.
        losses (int): Number of games a football team has lost.

        Note: Wins, Draws, and Losses should be positive integers.

    Returns:
        int | str:
            Total points achieved by a football team so far.
    '''

    if isinstance(wins, int) and wins >= 0 and isinstance(draws, int) and draws >= 0 and isinstance(losses, int) and losses >= 0:
        return (wins * 3) + draws
    
    return 'Please enter valid data'

num_wins, num_draws, num_losses = 5, 0, 2
print(f'Number of Wins: {num_wins}')
print(f'Number of Draws: {num_draws}')
print(f'Number of Losses: {num_losses}')
print(f'Total Points: {football_points(num_wins, num_draws, num_losses)}')
```

## Problem 16: The Farm Problem
A farmer is asking you to tell him how many legs can be counted among all his animals. The farmer breeds three species:
- Chickens = 2 legs
- Cows = 4 legs
- Pigs = 4 legs

```python
# the farm problem
def count_legs(num_chickens: int = 0, num_cows: int = 0, num_pigs: int = 0) -> int | str:
    '''
    This function calculates the number of legs on a farm.
    Chicken = 2 Legs
    Cows = 4 Legs
    Pigs = 4 Legs

    Args:
        num_chickens (int): Number of chickens on the farm
        num_cows (int): Number of cows on the farm
        num_pigs (int): Number of pigs on the farm
        Note: Number of chickens, cows, and pigs should be positive integers

    Return:
        int | str:
            Total number of legs on the farm
    '''

    if (
        isinstance(num_chickens, int) and num_chickens >= 0 and 
        isinstance(num_cows, int) and num_cows >=0 and 
        isinstance(num_pigs, int) and num_pigs >= 0
        ):
        return (num_chickens * 2) + (num_cows * 4) + (num_pigs) * 4
    
    return 'Please enter valid data'

num_chickens, num_cows, num_pigs = 5, 2, 8
print(f'Number of Chickens: {num_chickens}')
print(f'Number of Cows: {num_cows}')
print(f'Number of Pigs: {num_pigs}')
print(f'Total number of legs on the farm: {count_legs(num_chickens, num_cows, num_pigs)}')
```

## Problem 17: Convert Hours & Minutes Into Seconds

Write a function that takes two integers (hours, minutes), converts them to seconds, and adds them.

```python
# convert hours and minutes into seconds
def hours_minutes(hours: int = 0, minutes: int = 0) -> int | str:
    '''
    This function converts hours and minutes into seconds and returns the total seconds.

    Args:
        hours (int): Total number of hours
        minutes (int): Total number of minutes
        Note: Hours and Minutes should be positive integer values

    Returns:
        int | str:
            Total number of seconds
    '''

    if isinstance(hours, int) and hours >= 0 and isinstance(minutes, int) and minutes >= 0:
        return (hours * 60 * 60) + (minutes * 60)
    
    return 'Please enter valid data'

num_hours, num_minutes = 1, 3
print(f'Total number of hours: {num_hours}')
print(f'Total number of minutes: {num_minutes}')
print(f'Total number of seconds: {hours_minutes(num_hours, num_minutes)}')
```

## Problem 18: Basketball Points

You are counting points for a basketball game, given the amount of 3-pointers scored and 2-pointers scored, find the final points for the team and return that value.

```python
# basketball points
def basketball_points(three_pointers: int = 0, two_pointers: int = 0) -> int | str:
    '''
    This function calculates the score of a basketball game based on the 3-pointers and 2-pointers shot.

    Args:
        three_pointers (int): Number of 3-pointers shot.
        two_pointers (int): Number of 2-pointers shot.
        Note: 3-pointers and 2-pointers should be positive integer values.

    Returns:
        int | str:
            Basketball game score.
    '''

    if isinstance(three_pointers, int) and three_pointers >= 0 and isinstance(two_pointers, int) and two_pointers >= 0:
        return (3 * three_pointers) + (2 * two_pointers)
    
    return 'Please enter valid data'

three_pointers, two_pointers = 38, 8
print(f'Number of 3-pointers: {three_pointers}')
print(f'Number of 2-pointers: {two_pointers}')
print(f'Game score: {basketball_points(three_pointers, two_pointers)}')
```

## Problem 20: Find The Largest & The Smallest Number In A List

Create a function that takes a list of numbers and returns the largest and the smallest numbers in the list.

```python
# largest and smallest number
def largest_smallest(numbers: list[int | float]) -> tuple | str:
    '''
    This function returns the largest and the smallest numbers from a list.

    Args:
        numbers(list[int | float]): List of numbers
        Note: List should not be empty.

    Returns:
        tuple | str:
            Largest and smallest numbers from the list.
    '''

    if isinstance(numbers, (list, set, tuple)) and numbers:
        return max(numbers), min(numbers)
    
    return 'Please enter valid data'

list_of_numbers = [300, 200, 600, 150]
print(f'List of Numbers: {list_of_numbers}')

largest_number, smallest_number = largest_smallest(list_of_numbers)
print(f'\nLargest Number: {largest_number}\t\tSmallest Number: {smallest_number}')
```

## Problem 22: Create A Function To Concatenate Two Lists

Create a function to concatenate two lists.

```python
# concatenate lists
def concatenate_lists(list_1: list, list_2: list) -> list | str:
    '''
    This function concatenates two lists.

    Args:
        list_1 (list): First list
        list_2 (list): Second list

    Returns:
        list | str:
            Concatenated list
    '''

    if isinstance(list_1, list) and list_1 and isinstance(list_2, list) and list_2:
        list_1.extend(list_2)
        return list_1
    
    return 'Please enter valid data'

list_1, list_2 = [1, 3, 5], [2, 6, 8]

print(f'List 1: {list_1}')
print(f'List 2: {list_2}')
print(f'\nConcatenated list: {concatenate_lists(list_1, list_2)}')
```

## Problem 23: Less Than 100?

```python
# less than 100?
def less_than_100(num_1: int | float, num_2: int | float) -> bool | str:
    '''
    This function returns True if sum of two numbers is less than 100, otherwise returns False.

    Args:
        num_1 (int | float): First numerical value.
        num_2 (int | float): Second numerical value.

    Returns:
        bool | str:
            Boolean result.
    '''

    if isinstance(num_1, (int, float)) and isinstance(num_2, (int, float)):
        return (num_1 + num_2) < 100
    
    return 'Please enter valid data.'

number_1, number_2 = 83, 34
print(f'Is sum of {number_1} and {number_2} less than 100?: {less_than_100(number_1, number_2)}')
```

## Problem 25: Two Makes Ten

Create a function that takes two arguments - both being integers. Return True if one of them is 10 or if their sum is 10.

```python
# two makes ten
def two_makes_ten(num_1: int, num_2: int) -> bool:
    '''
    This function checks if the sum of two numbers is equal to ten or if one of the numbers is equal to ten.

    Args:
        num_1 (int): First integer value
        num_2 (int): Second integer value
        Note: Both values must be integer values

    Returns:
        bool:
            Boolean value based on evaluation.
    '''

    if isinstance(num_1, int) and isinstance(num_2, int):
        if (num_1 == 10) or (num_2 == 10) or (num_1 + num_2 == 10):
            return True
        
    return False

number_1, number_2 = 9, 10
print(f'Does sum of {number_1} and {number_2} equals to 10 or is any value equals to 10?: {two_makes_ten(number_1, number_2)}')
```

## Problem 26: Profitable Gamble

Create a function that takes three arguments prob, prize, pay and returns True if prob * prize > pay; otherwise return False.

```python
# profitable gamble
def profitable_gamble(prob: float = 0, prize: int = 0, pay: int = 0) -> bool:
    '''
    This function calculates if one should make a gamble or not. 

    Args:
        prob (float): Probability of winning a gamble
        prize (int): Prize money if one wins a gamble
        pay (int): Amount placed as a bet.

    Returns:
        bool:
            True if one should place a gamble, False otherwise.
    '''

    if isinstance(prob, float) and prob >= 0 and isinstance(prize, int) and prize >= 0 and isinstance(pay, int) and pay >= 0:
        return (prob * prize) > pay
    
    return False

probability, prize, pay = 0.2, 50, 9
print(f'Probability of winning a gamble: {probability * 100}%')
print(f'Prize money: {prize}')
print(f'Bet amount: {pay}')
print(f'\nShould one place the gamble?: {profitable_gamble(probability, prize, pay)}')
```

## Problem 29: Divides Evenly

Given two integers, a and b, return True if a can be divided evenly by b. Return False otherwise.

```python
# divides evenly
def divides_evenly(a: int, b: int) -> bool:
    '''
    This function checks if a number can be divided evenly by another or not.

    Args:
        a (int): First integer value
        b (int): Second integer value

    Returns:
        bool | str: 
            True if numbers are evenly divisible, False otherwise.
            For invalid data or 0 division, prints a message.
    '''

    if isinstance(a, int) and isinstance(b, int) and b != 0:
        return (a % b) == 0
    
    return 'Please enter valid data'

num_1, num_2 = 98, 7
print(f'Is {num_1} perfectly divisible by {num_2}?: {divides_evenly(num_1, num_2)}')
```

## Problem 30: Let's Fuel Up

A vehicle needs 10 times the amount of fuel than the distance it travels. However, it must always carry a minimum of 100 fuel before setting off.

Create a function which calculates the amount of fuel it needs, given the distance.

```python
# let's fuel up
def fuel_up(distance: int | float) -> int | float | str:
    '''
    This function calculates the fuel required for a journey.

    Args:
        distance (int | float): Distance to be covered.
        Note: Distance should always be greater than 0.

    Returns:
        int | float | str:
        Returns the fuel required to cover the journey.
        In case of invalid distance, returns a message.
    '''

    if isinstance(distance, (int, float)) and distance >= 10:
        return distance * 10
    elif isinstance(distance, (int, float)) and distance >= 1:
        return 100
    
    return 'Please enter valid data.'

distance = 23.5
print(f'Distance to be covered: {distance} units.')
print(f'Fuel required: {fuel_up(distance)} units.')
```
