## Stuttering Function

Write a function that stutters a word as if someone is struggling to read it. The first two letters are repeated twice with an ellipsis ... and space after each, and then the word is pronounced with a question mark ?.

```python
# stuttering function
def stutter(word: str) -> str:
    '''
    This function mimics stuttering.

    Args:
        word (str): A word that's to be converted into stuttering format.
    
    Returns:
        str: Word formatted in stuttering format.
    '''

    try:
        len(word) >= 2
        return f'{word[:2]}.. {word[:2].lower()}.. {word.lower()}?'

    except:
        return f'word?'

my_word = 'Outstanding'
print(f'{stutter(my_word)}')
```

## Discounted Price

Create a function that takes two arguments: the original price and the discount percentage and returns the final price after the discount.

```python
# find the discount

def discounted_price(price: int | float, discount: int | float) -> float | str:
    '''
    This function returns the discounted price of a product.

    Args:
        price (int | float): Retail price of the product.
        discount (int | float): Discount percent offered on the product.

    Returns:
        float | str: Returns the discounted price of the product. In case of incorrect input, raises error. 
    '''

    try:
        return price - (price * discount / 100)

    except TypeError:
        return 'Please enter valid data.'

retail_price = 1330
discount_offered = 30

print(f'Original price: {retail_price}')
print(f'Discount offered: {discount_offered}%')
print(f'Discounted price: {discounted_price(retail_price, discount_offered)}')
```

## Radians To Degrees

Create a function that takes an angle in radians and returns the corresponding angle in degrees rounded to one decimal place.

```python
# radians to degrees
def rad_deg(radians: int) -> float | str:
    '''
    This function converts radians to degrees.

    Args:
        radians (int): Angle measurement in radians.

    Returns:
        float | str: Angle measurement in degrees rounded to 1 decimal place. Raises error in case of incorrect input.
    '''

    from math import pi

    try:
        return round((180 / pi) * radians, 1)

    except TypeError:
        return 'Please enter valid data'

measurement_radians = 1
print(f'Measurement in radians: {measurement_radians}')
print(f'Measurement in degrees: {rad_deg(measurement_radians)}')
```

## Circle Or Square

Given the radius of a circle and the area of a square, return True if the circumference of the circle is greater than the square's perimeter and False if the square's perimeter is greater than the circumference of the circle.

```python
# circle or square
def circle_square(radius_cm: int, area_square: int) -> bool | str:
    '''
    This function returns a boolean result if the circumference of the circle is greater than the square's perimeter or not.

    Args:
        radius_cm (int): Radius of circle, in cm.
        area_square (int): Area of circle, in sq. cm.

    Returns:
        bool | str: 
            Return a boolean result if the circumference of the circle is greater than the square's perimeter or not.
            In case of incorrect data, returns error message.
    '''

    try:
        from math import pi, sqrt

        circle_circumference = 2 * pi * radius_cm 
        square_side = sqrt(area_square)
        square_perimeter = 4 * square_side

        return circle_circumference > square_perimeter

    except TypeError:
        return 'Please enter valid data.'

    except ValueError:
        return 'Numbers should be non-negative.'
    
radius, area = 16, 625

print(f'Radius of circle in cm: {radius}')
print(f'Area of square in sq. cm.: {area}')
print(f'\nIs circumference of the circle is greater than the square\'s perimeter?: {circle_square(radius, area)}')
```

## Curzon Number

Establish if a given integer num is a Curzon number. If 1 plus 2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon number.

Given a non-negative integer num, implement a function that returns True if num is a Curzon number, or False otherwise.

```python
# curzon numbers
def is_curzon(num: int) -> bool | str:
    '''
    This function determines if the number is a Curzon or not.

    Args:
        num (int): A non-negative integer value to check.

    Returns:
        bool | str:
            Returns True if the number is a Curzon number,
            False if it is not, or an error message if the
            input is invalid.
    '''

    try:
        if num < 0:
            raise ValueError('Number must be non-negative.')
        
        return (2 ** num + 1) % (2 * num + 1) == 0
    
    except TypeError:
        return 'Please enter valid data.'
    
    except ValueError:
        return 'Number must be non-negative.'
    
number = 10
print(f'Is {number} a Curzon?: {is_curzon(number)}')
```
